## Imbalanced Data Challenge

### 0 - Importing libraries & Loading data

```python
import pandas as pd         
import numpy as np          
import seaborn as sns      
import matplotlib.pyplot as plt 
```

```python
credit_raw = pd.read_csv("creditcard.csv")
credit_raw.head(3)
```

### 1 - Exploratory Data Analysis

```python
credit_raw.info()
```

```python
credit_raw.describe()
```

```python
print("\nTarget Distribution (Class):")
print(credit_raw['Class'].value_counts(normalize=True))

plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=credit_raw, palette='Set2')
plt.title('Distribution of the Target Variable (Class)')
plt.xlabel('Class (0 = Not Fraud, 1 = Fraud)')
plt.ylabel('Frequency')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 6))

sns.histplot(credit_raw['Amount'], bins=50, color='skyblue', kde=True, ax=ax[0])
ax[0].set_title('Distribution of Transaction Amounts with KDE')
ax[0].set_xlabel('Transaction Amount')
ax[0].set_ylabel('Frequency')

sns.histplot(credit_raw['Time'], bins=50, color='orange', kde=True, ax=ax[1])
ax[1].set_title('Distribution of Transaction Time with KDE')
ax[1].set_xlabel('Transaction Time (in seconds)')
ax[1].set_ylabel('Frequency')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Class', y='Amount', data=credit_raw, palette='Set3')
plt.title('Boxplot of Transaction Amount by Class')
plt.xlabel('Class (0 = Not Fraud, 1 = Fraud)')
plt.ylabel('Transaction Amount')
plt.show()
```
